### Combine Data in Pandas

```python
f = [2, 5, 6, 2, 1, 0, 7, 4]
s =[3, 10, 5, 0, 1, -1, 8, 15]
fs = pd.Series(f)
ss = pd.Series(s)
rs = fs.combine(ss, lambda a, b: a if a>b else b)
rs
```
